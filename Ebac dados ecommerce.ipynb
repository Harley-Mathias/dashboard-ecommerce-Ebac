```python
# Importando as bibliotecas que serao usadas
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
```

```python
#primeira leitura dos dados
df = pd.read_csv(
    r"C:\Users\Pipop\OneDrive\Ambiente de Trabalho\Ebac Python\ecommerce_estatistica.csv"
)
df.head()
```

```python
#exploração basiquinha
df.info()
df.describe()
```

```python
# Gráfico 1 – Distribuição de Preços

fig1 = px.histogram(
    df,
    x="Preço",
    nbins=30,
    title="Distribuição de Preços",
    labels={'Preço': 'Preço (R$)'},
    template="plotly_white"
)
fig1.show()
```

```python
## Gráfico 2 – Relação entre Preço e Nota

fig2 = px.scatter(
    df,
    x="Preço",
    y="Nota",
    color="Gênero",
    hover_data=['Título'],
    title="Relação Preço vs Nota",
    template="plotly_white"
)
fig2.show()
```

Este gráfico analisa a relação entre o preço dos produtos e a nota dada pelos consumidores.

```python
## Gráfico 3 – Mapa de Calor de Correlação

corr = df[['Nota', 'N_Avaliações', 'Desconto', 'Preço', 'Qtd_Vendidos_Cod']].corr()

fig3 = px.imshow(
    corr,
    text_auto=True,
    aspect="auto",
    color_continuous_scale='RdBu_r',
    title="Matriz de Correlação"
)
fig3.show()
```

O mapa de calor mostra a correlação entre variáveis numéricas do conjunto de dados, indicando quais tem uma relação forte ou fraca.

```python
"""Fiz esse tratamento pra eliminar algumas inconsistências que achei estranho. Para a analise de temporadas retornava alguns valores como ano (2021) que não faziam sentido para uma analise de estações"""

temporadas_validas = ['primavera/verão', 'outono/inverno', 'não definido']

df_temp = (
    df[df['Temporada'].isin(temporadas_validas)]
    ['Temporada']
    .value_counts()
    .reset_index()
)

df_temp.columns = ['Temporada', 'Contagem']
```

```python
## Gráfico 4 – Produtos por Temporada

fig4 = px.bar(
    df_temp,
    x='Temporada',
    y='Contagem',
    color='Temporada',
    title="Quantidade de Produtos por Temporada (Dados Filtrados)"
)
fig4.show()

```

Este gráfico apresenta a quantidade de produtos disponíveis em cada temporada.

```python
## Gráfico 5 – Distribuição por Gênero
fig5 = px.pie(
    df,
    names='Gênero',
    title='Distribuição de Produtos por Gênero'
)
fig5.show()
```

Este gráfico mostra a proporção de produtos por gênero do cliente.

```python
## Gráfico 6 – Densidade das Notas

hist_data = [df['Nota'].dropna()]
group_labels = ['Nota']

fig6 = ff.create_distplot(
    hist_data,
    group_labels,
    show_hist=False,
    show_rug=False
)

fig6.update_layout(title_text='Curva de Densidade das Notas')
fig6.show()
```

A curva de densidade permite visualizar a concentração relativa das notas dos produtos.

```python
## Gráfico 7 – Regressão entre Desconto e Vendas

fig7 = px.scatter(
    df,
    x="Desconto",
    y="Qtd_Vendidos_Cod",
    trendline="ols",
    title="Regressão: Impacto do Desconto nas Vendas",
    labels={
        'Qtd_Vendidos_Cod': 'Unidades Vendidas',
        'Desconto': 'Desconto (%)'
    }
)
fig7.show()
```

Este gráfico analisa o impacto do desconto na quantidade de vendas.
